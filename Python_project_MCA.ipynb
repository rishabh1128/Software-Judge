### **Importing section_dict from drive**

```python
from google.colab import drive
import json

drive.mount('/content/drive')
section_dict = dict()

with open('/content/drive/MyDrive/section_dict.json', 'r', encoding = 'utf-8') as file:
  section_dict = json.load(file)
```

### **Creating model and performing semantic search**

```python

from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')#msmarco-distilbert-base-v4')
corpus = []

for v in section_dict.values():
  corpus.append(v)

corpus_embedding = model.encode(corpus, convert_to_tensor=True, show_progress_bar=True, batch_size=32)
#'X was caught being involved in the production of fake currency.','X tried to stab Y to death.'Man raped a 6 year old girl.'
query = ['Man raped a 6 years old girl.']

query_embedding = model.encode(query, convert_to_tensor=True)

total_hits = util.semantic_search(query_embedding, corpus_embedding, top_k = 10)

for i,hits in enumerate(total_hits):
  print()
  print(i+1)
  print("The accused may be charged under the following sections:\n")
  for hit in hits:
      
      desc = corpus[hit['corpus_id']]
      for k,v in section_dict.items():
        if v==desc:
          print(k + ' : \n' + v + '\n', 'Score: {:.4f}'.format(hit['score']))
```
